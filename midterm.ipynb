```python
def addMatrix(s, t) :
    r = []
    for i in range(0, len(s)) :
        r1 = []
        for j in range(0, len(s[i])) :
            r1.append(s[i][j] + t[i][j])
        r.append(r1)
    return r

s = [[1, 2, 3], [4, 5, 6]]
t = [[1, 2, 3], [4, 5, 6]]
print(addMatrix(s, t))
```

```python
# Join
left = pd.DataFrame({'key':['foo', 'foo'], 'lval':[1, 2]})
print(left, '\n')

right = pd.DataFrame({'key':['foo', 'foo'], 'rval':[4, 5]})
print(right, '\n')

# merge : DataFrame 합치기
# pd.merge(left, right, on = '기준열', how = '조인방식')
print(pd.merge(left, right, on='key'))  # catesian product, 순서쌍
```

# Pandas 실습 과제 내용

```python
# csv 파일을 불러온 후 각 행 (학생별) 의 총합과 평균을 계산하세요
f = open('grade.csv', 'r')
lines = f.readlines()
f.close()

line_1 = lines[0].strip().split(',')
data = [line.strip().split(',') for line in lines[1:]]

df = pd.DataFrame(data, columns=line_1)
df.iloc[:, 1:] = df.iloc[:, 1:].astype(int)
df['Sum'] = df.iloc[:, 1:].sum(axis=1)
df['Mean'] = df.iloc[:, 1:-1].mean(axis=1)

print(df)
```

```python
# df 생성 / 상위 3개 행 출력 / 'Name', 'Salary'만 포함하는 df 생성 / Age >= 30 직원들만 선택
df_salary = pd.DataFrame([
    [1, 'Alice', 25, 50000],
    [2, 'Bob', 30, 60000],
    [3, 'Charlie', 35, 70000],
    [4, 'David', 40, 80000],
    [5, 'Eva', 45, 90000]
], columns=['ID', 'Name', 'Age', 'Salary'])

print(df_salary, '\n')
print(df_salary.iloc[:3], '\n')

df_name_salary = df_salary[['Name', 'Salary']]
print(df_name_salary, '\n')
print(df_salary[df_salary['Age'] >= 30])
```

```python
# df 생성 / 컬럼별 결측치 개수 확인 / 데이터 전체의 결측치 개수 확인
# Age의 결측치는 평균값으로 채움 / Name의 결측치 Unknown으로 채움 / Salary의 결측치 0으로 채움
df_salary = pd.DataFrame([
    [1, 'Alice', 25, 50000],
    [2, 'Bob', np.nan, 60000],
    [3, np.nan, 35, np.nan],
    [4, 'David', 40, 80000],
    [5, 'Eva', np.nan, 90000]
], columns=['ID', 'Name', 'Age', 'Salary'])

print(df_salary, '\n')
print(df_salary.isna().sum(), '\n')
print(df_salary.isna().sum().sum(), '\n')

df_salary['Age'].fillna(value = df_salary['Age'].mean())
df_salary['Name'].fillna(value = 'Unknown')
df_salary['Salary'].fillna(0)
```

```python
# df에서 Department별 평균 Salary를 계산
df = pd.DataFrame([
    [1, 'Alice', 'HR', 50000],
    [2, 'Bob', 'IT', 60000],
    [3, 'Charlie', 'IT', 70000],
    [4, 'David', 'HR', 80000],
    [5, 'Eva', 'IT', 90000]
], columns=['ID', 'Name', 'Department', 'Salary'])

print(df)

average_salary = df.groupby('Department')['Salary'].mean()
average_salary
```

```python
# df 두개 병합 : Inner Join / Left Join / Outer Join
df1 = pd.DataFrame([
    [1, 'Alice', 25],
    [2, 'Bob', 30],
    [3, 'Charlie', 35]
], columns=['ID', 'Name', 'Age'])

df2 = pd.DataFrame([
    [1, 50000],
    [2, 60000],
    [4, 80000]
], columns=['ID', 'Salary'])

inner_join = pd.merge(df1, df2, on='ID', how='inner')
print(inner_join, '\n')

left_join = pd.merge(df1, df2, on='ID', how='left')
print(left_join, '\n')

outer_join = pd.merge(df1, df2, on='ID', how='outer')
print(outer_join)
```

```python
# df에 Score가 50이상이면 Pass, 그렇지 않으면 Fail인 새로운 컬럼 Result 추가
df = pd.DataFrame([
    ['Alice', 85],
    ['Bob', 40],
    ['Charlie', 70],
    ['David', 90],
    ['Eva', 30]
], columns=['Name', 'Score'])

df['Result'] = df['Score'].apply(lambda x: 'Pass' if x >= 50 else 'Fail')

print(df)
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
import pandas as pd

driver = webdriver.Chrome()

# url 접근
url_CNN = 'https://edition.cnn.com/search?q=trump&from=0&size=10&page=1&sort=newest&types=all&section='
driver.get(url_CNN)
time.sleep(2)

# 스크롤 3번 내리기
for i in range(3) :
    driver.execute_script("window.scrollTo(0, document.documentElement.scrollHeight);")
    time.sleep(1)

# 요약기사 내용 수집
news_list = []
headline = driver.find_elements(By.CSS_SELECTOR, ".container__headline-text")
description = driver.find_elements(By.CSS_SELECTOR, ".container__description")
for i in range(len(headline)) :
    news_list.append({
        "Title" : headline[i].text.strip(),
        "Summary" : description[i].text.strip()
    })

driver.quit()

# df 저장
news_df = pd.DataFrame(news_list)
news_df
```

```python
# 형태소 분석
from nltk.tokenize import word_tokenize

all_text = " ".join(news_df["Title"].tolist())
nouns = word_tokenize(all_text)

# wordcloud 그리기
from wordcloud import WordCloud
import matplotlib.pyplot as plt

wordcloud = WordCloud(
    font_path='/System/Library/Fonts/Supplemental/Academy Engraved LET Fonts.ttf',
    background_color="black"
).generate(all_text)

plt.figure(figsize=(10,6))
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```

```python
# 빈도 히스토그램으로 표현
from collections import Counter

count = Counter(nouns)
top_20 = count.most_common(20)
words_top, counts_top = [], []
for word, cnt in top_20 :
    words_top.append(word)
    counts_top.append(cnt)
    
plt.figure(figsize=(10,6))
plt.bar(words_top, counts_top)
plt.xticks(rotation=45)
plt.show()
```
