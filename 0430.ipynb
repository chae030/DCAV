```python
import yfinance as yf

ticker = '^GSPC'

gspc_data = yf.download(ticker, start="2023-01-01", end="2024-01-01")
```

```python
df = pd.DataFrame(gspc_data)
df.columns = df.columns.droplevel(1)
df.columns.name = None
df = df.reset_index()
df.head()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))

plt.plot(df['Close'], label='S&P 500 Close Price')
plt.plot(df['High'], label='S&P 500 High Price')
plt.plot(df['Low'], label='S&P 500 Low Price')
plt.plot(df['Open'], label='S&P 500 Open Price')

plt.title('S&P 500 Close Price (2023)', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Close Price (USD)', fontsize=12)

plt.grid(True)
plt.legend()
plt.show()
```
