```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
import time
from selenium.webdriver.common.keys import Keys
import pandas as pd

service = Service("/Users/parkchaeyeon/Desktop/GitHub/Data collection and visualization/chromedriver-mac-arm64/chromedriver")
driver = webdriver.Chrome(service = service)
```

```python
data_th = []

for i in range(1,4):
    url = "https://section.blog.naver.com/Search/Post.naver?pageNo="+str(i)+"&rangeType=ALL&orderBy=sim&keyword=%ED%83%84%ED%95%B5"
    driver.get(url)
    time.sleep(1)
    titles_th = driver.find_elements(By.CSS_SELECTOR, "span.title")
    authors_th = driver.find_elements(By.CSS_SELECTOR, "em.name_author")
    for i in range(len(titles_th)):
        title = titles_th[i].text.strip()
        author = authors_th[i].text.strip()
        data_th.append({
            "Title": title,
            "Author": author
        })

df_th = pd.DataFrame(data_th)
df_th
```

```python
keywords = ["박보검", "아이유", "문소리"]
data_actor = []

for i in keywords :
    url_actor = "https://section.blog.naver.com/Search/Post.naver?pageNo=1&rangeType=ALL&orderBy=sim&keyword="+i
    driver.get(url_actor)
    time.sleep(1)
    titles_actor = driver.find_elements(By.CSS_SELECTOR, ".title_post > .title")
    contents_actor = driver.find_elements(By.CSS_SELECTOR, ".desc > a.text")
    for i in range(len(titles_actor)) :
        data_actor.append({
            "Title" : titles_actor[i].text.strip(),
            "Content" : contents_actor[i].text.strip()
        })

df_actor = pd.DataFrame(data_actor)
df_actor
```

```python
keywords = ["박보검", "아이유", "문소리"]
data_ac = []

for k in keywords :
    for i in range(1, 4) :
        url_ac = "https://section.blog.naver.com/Search/Post.naver?"+str(i)+"&rangeType=ALL&orderBy=sim&keyword="+k
        driver.get(url_ac)
        time.sleep(1)
        titles_ac = driver.find_elements(By.CSS_SELECTOR, ".title_post > .title")
        contents_ac = driver.find_elements(By.CSS_SELECTOR, ".desc > a.text")
        for j in range(len(titles_ac)) :
            data_ac.append({
                "Actor" : k,
                "Title" : titles_ac[j].text.strip(),
                "Content" : contents_ac[j].text.strip()
            })
df_ac = pd.DataFrame(data_ac)
df_ac
```
