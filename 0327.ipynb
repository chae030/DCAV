```python
import requests
from bs4 import BeautifulSoup

url_finance_naver = "https://finance.naver.com/item/main.naver?code=005930"
headers = {'User-Agent' : 'Mozila/5.0'}

response = requests.get(url_finance_naver, headers = headers)
soup = BeautifulSoup(response.content, "html.parser", from_encoding="utf-8")

lwidth = soup.select_one(".gray > .lwidth > .strong > td > em")
print(lwidth.text)
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
```

```python
service = Service("/Users/parkchaeyeon/Desktop/GitHub/Data collection and visualization/chromedriver-mac-arm64/chromedriver")
driver = webdriver.Chrome(service=service)
```

```python
url_blog = "http://section.blog.naver.com/BlogHome.naver?directoryNo=0&currentPage=1&groupId=0"
driver.get(url_blog)
print(driver.page_source)
```

```python
blog_list = driver.find_elements(By.CSS_SELECTOR, ".info_post")

for blog in blog_list :
    id = blog.find_element(By.CSS_SELECTOR, ".name_author").text
    title = blog.find_element(By.CSS_SELECTOR, ".title_post").text
    text = blog.find_element(By.CSS_SELECTOR, ".text").text
    print(id, ' - ', title, ' - ', text)
```

```python
url_finance_naver = "https://finance.naver.com/item/main.naver?code=005930"
driver.get(url_finance_naver)
finance_info = driver.find_element(By.CSS_SELECTOR, "div.gray > table.lwidth > tbody > tr.strong > td > em").text
print(finance_info)
```

```python
url_melon = "https://www.melon.com/chart/index.htm"
driver.get(url_melon)
top_100 = driver.find_elements(By.CSS_SELECTOR, "div.service_list_song.type02.d_song_list > table > tbody > tr#lst50.lst50 > td > div.wrap.t_center > span.rank")
for i in top_100 : 
    print(i.text)
```
