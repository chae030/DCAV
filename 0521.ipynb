```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

import warnings
warnings.filterwarnings('ignore')
```

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

```python
tips.info()
```

```python
sns.catplot(x = 'day', y = 'total_bill', data = tips, height = 3,kind = 'swarm', palette = 'Set3')
```

```python
sns.catplot(x = 'day', y = 'total_bill', data = tips, kind = 'box', palette = 'Set3', hue = 'sex')
```
