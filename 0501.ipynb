```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.rcParams['font.family'] = ['']
```

```python
months = ['1m', '2m', '3m', '4m']

A = [0.5,0.4,0.6,0.7]
B = [0.2,0.1,0.3,0.2]

plt.figure(figsize=(8,6))

plt.plot(months, A, color = 'red', marker='o', label='group A')
plt.plot(months, B, color = 'blue', marker='s', label='group B')
plt.legend(loc='upper left')
plt.title('title')
plt.xlabel('months')
plt.ylabel('a')
plt.ylim(0,1)

plt.show()
```

```python
import yfinance as yf

ticker = '^GSPC'

gspc_data = yf.download(ticker, start="2020-01-01", end="2025-04-30")
```

```python
gspc_data
```

```python
ohlc = gspc_data[['Open', 'High', 'Low', 'Close']]
ohlc
```

```python
ax1 = ohlc.plot(figsize=(12,6))
```

```python
ax2 = volume.plot()
```
